```python
def getPeakGain(gain, lengths, p)
    maxGain = 0
    for i in range(1, len(gain))
        if(maxGain < gain[i]):
            maxGain = gain[i]
            lengthMaxGain = lengths[i-1]
    print("Player: "+ str(p) + " peak gain: " + str(maxGain) + " at length: " + str(lengthMaxGain))       
    
```
